```python
example_prompt = '''
A customer is randomly selected while shopping for laptops. Their annual income is $ 10,000.
While shopping, the customer sees a a Surface Laptop 3, Price:  $1000,
Processor: Intel Core i5, RAM: 8GB, Screen Size: 13.5in, SD: 128GB
The customer is asked, after they finish shopping: Did you purchase any laptop? If so, which one?
Customer: 
'''
```
